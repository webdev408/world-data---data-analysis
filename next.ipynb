```python
import pandas as pd
```

```python
df = pd.read_html(r"C:\Users\webde\Downloads\WV.1_Size_of_the_economy.xls")[0]
df.head()
```

```python
if isinstance(df.columns, pd.MultiIndex):
  df.columns = ['_'.join([str(i) for i in col if str(i) !='nan']) for col in df.columns.values]
```

```python
print(df.columns.tolist())
```

```python
renamed_df = df.rename(columns={
  'Unnamed: 0_level_0_Unnamed: 0_level_1_Unnamed: 0_level_2_Unnamed: 0_level_3': 'country',
  'Population_Unnamed: 1_level_1_millions_2024': 'population',
  'Gross national income, Atlas method_Unnamed: 4_level_1_$ billions_2024': 'GNI (billions)',
  'Purchasing power parity gross national income_Unnamed: 6_level_1_$ billions_2024': 'PPPGni (billions)',
  'Gross domestic product_Unnamed: 8_level_1_% growth_2024': 'gdp_growth_rate'
})
```

```python
new_df = renamed_df[['country', 'population', 'GNI (billions)', 'PPPGni (billions)', 'gdp_growth_rate']].copy()
print(new_df.head())
```

```python
for idx, country in new_df['country'].items():
  print(idx, country)
```

```python
new_df = new_df.loc[0:214]
print(new_df)
```

```python
cols = ['population', 'GNI (billions)', 'PPPGni (billions)', 'gdp_growth_rate']
new_df[cols] = new_df[cols].apply(pd.to_numeric, errors='coerce')
print(new_df[cols])
```

```python
print(new_df[cols].head().round(2))
```

```python
print(new_df.describe())
```

```python
new_df['gni percapita'] = new_df['GNI (billions)'] / new_df['population']
print(new_df)
```

```python
print(new_df.nlargest(15, 'gni percapita'))
```

```python
print(new_df.nsmallest(15, 'gni percapita'))
```

```python
south_asia = new_df[new_df['country'].isin(['Bangladesh', 'Bhutan', 'India', 'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka'])&
                    (new_df['gni percapita'] <= 4000)]
print(south_asia)
```
